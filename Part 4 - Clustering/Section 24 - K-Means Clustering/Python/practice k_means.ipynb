# K-Means Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans

scores = []

for i in range(1, 11):
    k = KMeans(n_clusters=i, random_state=42, init='k-means++').fit(X)
    scores.append(k.inertia_)
plt.plot( range(1,11),scores , color = 'blue')
plt.title('elbow method')
plt.xlabel('Num of clusters ')
plt.ylabel('wcss score')
plt.show()
```

## Training the K-Means model on the dataset

```python
k = KMeans(n_clusters=5, random_state=42, init='k-means++')
y_means = k.fit_predict(X)
```

## Visualising the clusters

```python
plt.scatter(X[y_means == 0,0],X[y_means == 0,1], s= 100, c='red', label='Cluster 1')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1], s= 100, c='blue', label='Cluster 2')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1], s= 100, c='black', label='Cluster 3')
plt.scatter(X[y_means == 3,0],X[y_means == 3,1], s= 100, c='cyan', label='Cluster 4')
plt.scatter(X[y_means == 4,0],X[y_means == 4,1], s= 100, c='green', label='Cluster 5')
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1], s=300, c='yellow', label='Centroids')
plt.title('costumer clusters')
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
